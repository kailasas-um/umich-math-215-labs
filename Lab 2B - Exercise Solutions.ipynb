### Exercise 2:

Looking at the heatmap for the graph of $f_3$, we expect that $f_3 \to 0$ as $(x,y) \to (0,0)$, regardless of the direction from which we approach $(0,0)$. To prove this, let's pick an arbitrary direction $\langle a, b \rangle$ and approach the origin from this direction along the parametric curve $\gamma(t) = \langle ta, tb \rangle$. We then have 

$$\lim_{t\to 0^{+}} f(ta, tb) = \lim_{t\to 0^{+}} \frac{t^2 |ab|}{t(|a|+|b|)} = \lim_{t\to 0^{+}} \frac{t |ab|}{|a|+|b|} = 0$$

as desired.

### Exercise 4:

The function $f(x,y) = x^y$ equals $A$ along the curve $x^y = A \implies y \ln(x) = \ln(A)$. Let's parametrize this curve as $\gamma(t) = \left\langle t, \frac{\ln(A)}{\ln(t)} \right\rangle$. Then, we want to calculate $\lim_{t\to 0^{+}} f(t, \gamma(t))$. Rather than calculating this limit directly, let's take the logarithm, calculate the limit in that setting, and then exponentiate the result: $$\lim_{t\to 0^{+}} \ln(f(\gamma(t)) = \lim_{t\to 0^{+}}  \frac{\ln(A)}{\ln(t)} \cdot \ln(t) = \ln(A)$$ Thus, $\lim_{t\to 0^{+}} f(\gamma(t)) = A$.
