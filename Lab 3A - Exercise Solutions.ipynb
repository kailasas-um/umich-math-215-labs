**Exercise 2:**

Points along the solution curve satisfy $$H = \frac{112-LW}{L+W} = L+W \implies (L+W)^2 = 112-LW$$ $$\implies W^2 + 3LW + L^2 - 112 = 0$$ Thus, using the quadratic formula $$W = \frac{-3L\pm \sqrt{5L^2 + 448}}{2}$$ We conclude that the intersection curve is parametrized by $L$ as 

$$\gamma(L) = \left(L, \frac{-3L \pm \sqrt{5L^2 + 448}}{2}, L +  \frac{-3L \pm \sqrt{5L^2 + 448}}{2} \right)$$

Plotting both of these solutions, we see that we can throw away the one with a minus sign: this corresponds to negative values of $L$ and $W$.

```matlab
lvec = [0:1:15]; wvec = [0:1:15];
[L,W]=meshgrid(lvec,wvec);

H1 = (112-L.*W)./(L+W);
H2 = L+W;

hold on;
colormap('hsv')

surf(L,W,H1)
surf(L,W,H2)
view(30,30)
grid on
xlabel('L'); ylabel('W'); zlabel('H')
plot3(lvec, (-3*lvec + (5*lvec.^2+4*112).^.5)/2, lvec+ (-3*lvec + (5*lvec.^2+4*112).^.5)/2, 'LineWidth',3)


figure()
hold on
plot(lvec, (-3*lvec + (5*lvec.^2+4*112).^.5)/2,'LineWidth',3)
contour(L,W, L.*W.*(L+W),50)
scatter(4.7329,4.7329)
colorbar
```
