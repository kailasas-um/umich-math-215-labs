## Review of Graphing Surfaces

The graph of a single variable function forms a curve in the $x$-$y$ plane. Given a function $f(x)$, the graph is defined by the formula $y=f(x)$: for each value of $x$, the formula specifies the _height_ ($y$-coordinate) of the corresponding point on the curve. Similarly, the graph of a two-variable function forms a surface in $x$-$y$-$z$ space. Given a function $f(x,y)$, the graph is defined by the formula $z = f(x,y)$. For each pair $(x,y)$, the formula tells us the height ($z$-coordinate) of the corresponding point on the surface.

In order to graph a surface, we need to decide the region of the $x$-$y$ plane over which we want to visualize it. Then, we divide this region up into a mesh, plot the corresponding surface point over each mesh point, and then connect the surface points. The Matlab functions `meshgrid` and `surf` facilitate this. Let's use them to graph the quadratic surface $$z = x^2 + 3xy-2y^2$$ over the $x$-$y$ region given by $-5 \le x \le 5$ and $-5\le y \le 5$.

```matlab
%First, we instantiate our x and y ranges
xvec = -5:.25:5;
yvec = -5:.25:5;

%Then, we feed these into meshgrid to obtain the coordinates of all possible points (x,y) with x in xvec and y in yvec
[X,Y] = meshgrid(xvec,yvec);
%X is a matrix containing x-coordinates, Y is a matrix containing the corresponding y-coordinates

%Creating matrix of heights over each point in the square mesh. Note use of .* rather than * for elementwise multiplication instead of matrix multiplication.
Z = X.^2 + 3*X.*Y - 2*Y.^2;

%Plot the surface with surf
surf(X,Y,Z);

%Prettify with axis labels
xlabel('x'); ylabel('y'); zlabel('z');

%Changes the viewing angle of the surface to azimuthal angle 50 degrees and elevation 25 degrees; look at the documentation for more details
view(50,25);

%Creates another view of the surface, as a heatmap
figure;
surf(X,Y,Z); view(2);
xlabel('x'); ylabel('y');
colorbar;
```

## Continuity

An intuitive definition of continuity: we say $f$ is continuous at $a$ if the value $f(a)$ can be unambiguously predicted from the "nearby" values of $f(x)$ for $x$ close, but not equal, to $a$. Let's look at an example of this in the single variable case. We consider the functions 

$$a(x) = x \sin(x)$$ 

and

$$b(x) = \lfloor x \rfloor \sin(x)$$ 

(where $\lfloor x \rfloor$ denotes the greatest integer less than or equal to $x$, i.e. the round down function. So, for example, $\lfloor 1.83 \rfloor = 1$, $\lfloor 3 \rfloor = 3$, and $\lfloor -8.2 \rfloor = -9$. This is called the *floor* function.)

```matlab
%We'll plot both functions over -10 <= x <= 10
xvec = -10:0.0001:10;

%Again, note the syntax carefully -- we are using .* for elementwise multiplication
avec = xvec .* sin(xvec);
bvec = floor(xvec) .* sin(xvec);

plot(xvec,avec,xvec,bvec)
legend('a(x)', 'b(x)')
xlabel('x'); ylabel('y'); title('Graphs of a(x) and b(x) (-10 <= x <= 10)');

%Let's zoom in on just the x-values 0 <= x <= 2.5

figure;
plot(xvec,avec,xvec,bvec)
xlim([0 2.5])
legend('a(x)', 'b(x)')
xlabel('x'); ylabel('y'); title('Graphs of a(x) and b(x) (0 <= x <= 2.5)');
```

Note that $a(1) = b(1) = \sin(1)$. We can predict the value of $a(1)$ from the values of $a(x)$ for $x$ slightly less than or slightly greater than $1$. On the other hand, $b(x) = 0$ for $x<1$; if we only knew the values of $b(x)$ for $x<1$, we would not guess that $b(1) = \sin(1)$. This appears on the graph as a "jump." Since $b(x) = \sin(x)$ for $1< x \le 2$, we _can_ predict $b(1)$ using these values (this is why people sometimes make the distinction between left- and right- continuous).

The word "unambiguous" in the definition of continuity indicates that if $f$ is continuous at $a$, we ought to be able to predict the correct value of $f(a)$ from nearby values of $f$, _regardless of the direction_ from which we approach $a$. This concept directly translates to the setting of multivariable functions. But while in dimension $1$, there are only $2$ directions from which to approach any point, in dimensions $>2$ there are infinitely many possible directions from which to approach. We only call a function continuous at $a$ if you can predict the value of $f(a)$ accurately by approaching $a$ from _any_ direction; if there are two approach directions that give differing predictions, then the function is not continuous at $a$.

In the mathematical lingo, if approaching $a$ from _any_ direction gives the same limiting value, we say that $$\lim_{x\to a} f(x) $$ exists. The condition for continuity at $a$ is that $$\lim_{x\to a} f(x) = f(a)$$


**Question:** Is $f(x,y) = x^2 + 3xy - 2y^2$ continuous at $(1,2)$?

Let's zoom in on the graph of the surface near this point. We see that all the values $f(x,y)$ for $(x,y) \approx (1,2)$ are close to $f(1,2) = -1$, so the function does appear to be continuous at $(1,2)$.

```matlab
surf(X,Y,Z);
hold on;
plot3( [1], [2], [-1], '.r', 'MarkerSize', 35)
axis( [0.5 1.5 1.5 2.5 -1.5 -0.5] );
xlabel('x'); ylabel('y'); zlabel('z');
view(10,20)
```

**Question:** Is the function 

$$
f(x,y) = \left\{
        \begin{array}{ll}
            \dfrac{\sin(xy)}{x^2 + y^2} & \quad (x,y) \neq (0,0) \\
            \dfrac{1}{2} & \quad(x,y) = (0,0)
        \end{array}
    \right.
$$

continuous at $(0,0)$? Let's first graph the surface to get a feel for things.



```matlab
xvec = -2:.05:2; yvec = -2:.05:2;
[X,Y] = meshgrid(xvec,yvec);
Z = sin(X.*Y)./(X.^2+Y.^2);

surf(X,Y,Z);
xlabel('x'); ylabel('y'); zlabel('z');

figure;
surf(X,Y,Z); view(2);
xlabel('x'); ylabel('y'); zlabel('z');
colorbar;
```

The heatmap suggests the function is **not** continuous at $(0,0)$. The value of $f(0,0)$ is $0.5$. From certain directions (the warm colored directions), it does appear that the nearby values of $f$ are approaching $0.5$. However, from other directions (the cool colored directions), the nearby values of $f$ are instead approaching $-0.5$. Let's explicitly see what's going on with the values of $f$ as we approach $(0,0)$ in a straight line starting either from $(1,1)$ or from $(1,-1)$. In other words, we're going to study the limits $$\lim_{t\to 0^{+}} f(t,t)$$ and $$\lim_{t\to 0^{+}} f(t,-t)$$ We expect these two directional limits to be different, and therefore the overall limit 

$$\lim_{(x,y) \to (0,0)} f(x,y)$$ 

fails to exist.

```matlab
x_11 = 0.01:.1:1;
y_11 = x_11;
x_1neg1 = x_11;
y_1neg1 = -x_1neg1;

z_11 = sin(x_11.*y_11)./(x_11.^2 + y_11.^2);
z_1neg1 = sin(x_1neg1.*y_1neg1)./(x_1neg1.^2 + y_1neg1.^2);

disp("Values of f from (1,1) direction:")
disp(z_11(1:10))

disp("Values of f from (1,-1) direction:")
disp(z_1neg1(1:10))
```

```matlab
hold on; grid on;
surf(X,Y,Z);
alpha 0.3;
plot3(x_11,y_11,z_11,'.r','MarkerSize',25);
plot3(x_1neg1,y_1neg1,z_1neg1,'.r','MarkerSize',25);
xlabel('x'); ylabel('y'); zlabel('z');
view(70,70);
```

# Curvature

Given a curve in space, we can calculate for each point on the curve a quantity called **curvature**. This quantity behaves as you might expect intuitively: roughly speaking, points where the curve turns sharply have high curvature and points where the curve turns slowly have lower curvature. 




We won't go into where the formula for curvature comes from right now; instead, we'll just calculate it for a specific curve and verify that it behaves the way we expect it to. Given a parametric curve $\gamma(t)$ in 3D space, the curvature of $\gamma$ at time $t$ is given by 

$$\kappa(t) := \frac{\|\gamma'(t) \times \gamma''(t) \|}{\|\gamma'(t)\|^{3}}$$

Let's calculate $\kappa$ for some points on the curve 

$$\gamma(t) = \langle \cos(t), \sin(t), \sin^{10} (t) \rangle$$

First, let's plot this curve to get a feel for how it looks. 

```matlab
%plot inline
t = -10:0.05:10;
t1 = [0 pi];
t2 = [pi/2 3*pi/2];
hold on;
plot3(cos(t),sin(t), sin(t).^10);
plot3(cos(t1),sin(t1),sin(t1).^10,'.r','MarkerSize',30)
plot3(cos(t2),sin(t2),sin(t2).^10,'.b','MarkerSize',30)
view(10,60);
xlabel('x'); ylabel('y'); zlabel('z'); grid on;
hold off;
figure;
hold on;
plot3(cos(t),sin(t), sin(t).^10);
plot3(cos(t1),sin(t1),sin(t1).^10,'.r','MarkerSize',30)
plot3(cos(t2),sin(t2),sin(t2).^10,'.b','MarkerSize',30)
hold off;
view(-50,60);
xlabel('x'); ylabel('y'); zlabel('z'); grid on;
```

From the plot above, it looks like the curve should have higher curvature at the blue dots ($t = 0, \pi$) than at the red dots ($t = \pi/2, 3\pi/2$). Now, let's calculate the curvature and see if the values are in agreement with our intuition.

```matlab
dt = 0.001;
t = -10:dt:10;

x = cos(t);
y = sin(t);
z = sin(t).^10;

%approximating the derivative using difference quotient
dxdt = diff(x)/dt; 
dydt = diff(y)/dt;
dzdt = diff(z)/dt;

%pad vectors so the length doesn't decrease by 1
dxdt = [0,dxdt]; dydt = [0,dydt]; dzdt = [0,dzdt];

%approximating the second derivative using difference quotient
ddxdtt = diff(dxdt)/dt;
ddydtt = diff(dydt)/dt;
ddzdtt = diff(dzdt)/dt;

%more padding
ddxdtt = [0,ddxdtt]; ddydtt = [0,ddydtt]; ddzdtt = [0,ddzdtt];

kappa = [];

for i=1:length(t)
    gammaprime = [dxdt(i), dydt(i), dzdt(i)];
    gammaprimeprime = [ddxdtt(i), ddydtt(i), ddzdtt(i)];
    curvature = norm(cross(gammaprime, gammaprimeprime))/norm(gammaprime)^3;
    kappa = [kappa, curvature];
end
```

```matlab
%plot inline
plot(t,kappa)
hold on
plot(0,0:.5:12,'.r')
plot(pi,0:.5:12,'.r')
plot(pi/2,0:.5:12,'.b')
plot(3*pi/2,0:.5:12,'.b')
xlim([-2*pi,2*pi])
xlabel('t'); ylabel('\kappa'); title('\kappa(t)')
```
