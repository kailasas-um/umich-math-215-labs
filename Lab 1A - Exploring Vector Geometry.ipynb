# Lab 1A: Exploring Vector Geometry

Last week, we mentioned two different ways to think about a **vector**:

* A vector is a *list of numbers*.
* A vector is a *special type of matrix with only one row or one column*.

In both these descriptions, the vector is thought of as a container holding numerical data. However, vectors have much richer lives than this; we can also interpret them *geometrically*, and this is a very powerful viewpoint. More precisely, we can add the following perspectives to our list:

* A vector of length $n$ represents a *point in $n$-dimensional space*.
* A vector of length $n$ represents a *displacement in $n$-dimensional space*.

You may have heard of the latter perspective in the more colloquial form of "a vector is an *arrow*." Let's clarify these ideas by looking at some examples in 2D space.

**Notation**: We'll denote $d$-dimensional space as $\mathbb{R}^d$.

## Vectors in $\mathbb{R}^2$

Consider the length 2 vectors $\mathbf{u} = (1,2)$ and $\mathbf{v} = (3,4)$. We can think of these two lists of numbers as specifying $x$ and $y$ coordinates for points in the plane. 

```matlab
hold on;
scatter(1,2,'rs','filled');
scatter(3,4,'bs','filled');
xlim([0.5,3.5]);
ylim([1.5,4.5]);
grid on;
text(1,2,'  \leftarrow u','FontSize',15);
text(2.65,4,'v \rightarrow', 'FontSize', 15);
```

We can also think of vectors as specifying *displacements* in the plane. You should visualize this as an arrow that begins at the origin and ends at the vector coordinates, as if to specify that an object began at the origin and was moved to the position indicated by the vector. 

```matlab
hold on;
plot([0,1],[0,2],'r');
plot([0,3],[0,4],'b');
xlim([0,3.5]);
ylim([0,4.5]);
text(1,2.2,'u','FontSize',15);
text(3,4.2,'v','FontSize',15);
grid on;
```

## Calculating angles and vector products

Recall that the **norm** of a vector $\mathbf{v} = (v_1, \cdots, v_n)$ is defined as

$$\| \mathbf{v} \| = \sqrt{v_{1}^{2} + v_{2}^{2} + \cdots + v_{n}^{2}} = \sqrt{\sum_{i=1}^{n} v_{i}^{2}}$$

and the **dot product** of two vectors $\mathbf{v} = (v_1, \cdots, v_n)$ and $\mathbf{w} = (w_1, \cdots, w_n)$ is defined as 

$$\mathbf{v} \cdot \mathbf{w} = v_1 w_1 + v_2 w_2 + \cdots + v_n w_n = \sum_{i=1}^{n} v_i w_i$$ 

The norm is also sometimes called the *length* of the vector, and measures the distance between the origin and the point the vector describes (alternatively, the magnitude of the displacement the vector describes). There is an important geometric relation between dot product and norm, which is useful for calculating the angle $\theta$ between a pair of vectors:

$$\mathbf{v} \cdot \mathbf{w} = \|\mathbf{v} \| \|\mathbf{w} \| \cos(\theta)$$ 

In particular, note that for $\theta = 0$ this reduces to the relation $\mathbf{v} \cdot \mathbf{v} = \|\mathbf{v}\|^{2}$, so the norm of a vector is the square root of the dot product of the vector with itself.

**Example**: Let's calculate the angle between $\mathbf{u} = (1,2)$ and $\mathbf{v} = (3,4)$.

```matlab
u = [1 2]
v = [3 4]
cos_theta = dot(u,v)/(norm(u)*norm(v))
theta = acos(cos_theta) % in radians
disp([num2str(theta*180/pi), ' degrees'])
```

Given a vector $\mathbf{v}$, we can obtain a **unit vector** by dividing $\mathbf{v}$ by its norm $\|\mathbf{v} \|$. This is a vector that points in the same direction as $\mathbf{v}$ but has length normalized to equal 1.

## Vector Projections

An extremely important concept for science and engineering is the **projection of one vector onto another vector**. If $\mathbf{v}$ and $\mathbf{w}$ are vectors, we denote the projection of $\mathbf{v}$ onto $\mathbf{w}$ as $\text{proj}_{\mathbf{w}}(\mathbf{v})$. You can think of $\text{proj}_{\mathbf{w}}(\mathbf{v})$ as the vector you would get if the whole 2D plane got squashed perpendicularly into the line defined by $\mathbf{w}$ (i.e. the line through the origin in the direction of $\mathbf{w}$). Another way to think about it is that $\text{proj}_{\mathbf{w}}(\mathbf{v})$ "forgets" everything about $\mathbf{v}$ except what is happening in direction $\mathbf{w}$. 

The projection has the two following key properties:

* $\text{proj}_{\mathbf{w}}(\mathbf{v})$ and $\mathbf{v}-\text{proj}_{\mathbf{w}}(\mathbf{v})$ are **orthogonal, a.k.a. perpendicular**, meaning the angle between these two vectors is $90^{\circ}$.
* $\text{proj}_{\mathbf{w}}(\mathbf{v})$ is proportional to $\mathbf{w}$.

These two properties together imply the following formula: $$\text{proj}_{\mathbf{w}}(\mathbf{v})  = \frac{\mathbf{v} \cdot \mathbf{w}}{\mathbf{w} \cdot \mathbf{w}} \mathbf{w}$$

```matlab
u = [1 2];
v = [3 4];
proj_u_onto_v = dot(u,v)/dot(v,v) * v
```

```matlab
hold on;
plot([0,u(1)],[0,u(2)],'r','DisplayName','u');
plot([0,v(1)],[0,v(2)],'b','DisplayName','v');
plot([0,proj_u_onto_v(1)],[0,proj_u_onto_v(2)],'g','DisplayName','proj_{v}(u)');
plot([proj_u_onto_v(1), u(1)],[proj_u_onto_v(2),u(2)],'m', 'DisplayName', 'u-proj_{v}(u)')

xlim([0,3.5]);
ylim([0,4.5]);
legend();
grid on;
```

In the plot above, the vectors $\mathbf{u}$ and $\mathbf{u} - \text{proj}_{\mathbf{v}}(\mathbf{u})$ are supposed to be perpendicular, but they don't appear to be. What's gone wrong? Look carefully at the axes: the units are different! We can fix this with the command `axis equal`.

```matlab
hold on;
plot([0,u(1)],[0,u(2)],'r','DisplayName','u');
plot([0,v(1)],[0,v(2)],'b','DisplayName','v');
plot([0,proj_u_onto_v(1)],[0,proj_u_onto_v(2)],'g','DisplayName','proj_{v}(u)');
plot([proj_u_onto_v(1), u(1)],[proj_u_onto_v(2),u(2)],'m', 'DisplayName', 'u-proj_{v}(u)')

axis equal;
xlim([0,3.5]);
ylim([0,4.5]);
legend();
grid on;
```

## Parametric Curves

Imagine there is a particle moving around in $\mathbb{R}^2$. We can describe its trajectory using two functions $x(t)$ and $y(t)$, which describe the $x$-coordinate and $y$-coordinate of the particle at time $t$, respectively. Such a pair of functions describes a **parametric curve**. Another way to think about this situation is that we have a *vector valued* function $\gamma(t) = (x(t), y(t))$ which, for each time $t$, yields a vector describing the position of the particle.

This same concept extends to parametric curves in $\mathbb{R}^3$, defined a vector valued function of the form $\gamma(t) = (x(t), y(t), z(t))$. In fact, an example of a parametric curve in $\mathbb{R}^3$ above, when we plotted the one parameter family of vectors.

**Example**: The simplest example of a parametric curve in $\mathbb{R}^2$ is given by $x(t) = \cos(t)$ and $y(t) = \sin(t)$. Since $\sin^2 (t) + \cos^2 (t) = 1$, this particle's path is constrained to lie on the unit circle. In fact, as $t$ ranges from $0$ to $2\pi$, this path traces out the unit circle exactly.

```matlab
t = 0:0.1:2*pi;
x = sin(t);
y = cos(t);

plot(x,y,'--b'); % -- makes dotted line
axis equal;
```

**Example**: Next, consider the pair of parametric curves $\mathcal{C}_1$ and $\mathcal{C}_2$ given by the equations

$$\mathcal{C}_1 : x(t) = \sin^3 (t), y(t) = \cos^3 (3t)$$ $$\mathcal{C}_2: x(t) = \sin^3 (3t), y(t) = \cos^3 (t)$$

Let's plot both of these curves.

```matlab
t = 0:.01:2*pi;
x1 = sin(t).^3;
x2 = sin(3*t).^3;
y1 = cos(3*t).^3;
y2 = cos(t).^3;

plot(x1, y1, 'm', x2, y2, 'g')
```
